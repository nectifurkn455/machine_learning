```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('/home/necati/machine_learning/Kaggle/Kaggle_datasets/trainHousePrice.csv')
test = pd.read_csv('/home/necati/machine_learning/Kaggle/Kaggle_datasets/testHousePrice.csv')
```

```python
train.isnull().sum()
```

```python
train.fillna(train.median(numeric_only=True), inplace=True)
```

```python
for col in train.select_dtypes(include=['object']):
    train[col].fillna(train[col].mode()[0], inplace=True)
```

```python
label_encoder = LabelEncoder()
for col in train.select_dtypes(include=['object']):
    train[col] = label_encoder.fit_transform(train[col])
    test[col] = label_encoder.fit_transform(test[col])
```

```python
train.info()
```

```python
y = train['SalePrice']
X = train.drop(columns=['Id', 'SalePrice'])
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
model1 = RandomForestRegressor(n_estimators=100, random_state=42)
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
rmse
```

```python
test_pred = model1.predict(test.drop(columns=['Id']))
```

```python
submission = pd.DataFrame({'Id': test['Id'],
                           'SalePrice': test_pred})
submission.to_csv('HousePrice_submission.csv', index=False)
```

# Gradient Boosting Regressor Model 

```python
model2 = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3)
model2.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_val)
rmse2 = root_mean_squared_error(y_val, y_pred2)
rmse2
```

```python
test.fillna(test.median(numeric_only=True), inplace=True)
for col in test.select_dtypes(include=['object']):
    test[col].fillna(train[col].mode()[0], inplace=True)
```

```python
test_pred2 = model2.predict(test.drop(columns = ['Id']))
```

```python
submission2 = pd.DataFrame({'Id': test['Id'], 
                            'SalePrice': test_pred2})
submission2.to_csv('HousePrice_submission2.csv', index=False)
```

# HyperParameter Tuning

```python
param_grid = {'n_estimators' : [100, 200, 300],
              'max_depth' : [None, 10, 20],
              'min_samples_split' : [2, 5, 10],
              'min_samples_leaf' : [1, 2, 4]}
grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
```

```python
model3 = RandomForestRegressor(random_state=42, max_depth=10, min_samples_leaf=2, min_samples_split=2, n_estimators=200)
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_val)
rmse3 = root_mean_squared_error(y_pred3, y_val)
rmse3
```

```python
param_grid2 = {'n_estimators' : [100, 200, 300, 400],
               'min_samples_split' : [2, 5, 10],
               'min_samples_leaf': [1, 2, 4]}
grid_search2 = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid2, cv=3, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_search2.fit(X_train, y_train)
print(grid_search2.best_params_)
```

```python
model4 = GradientBoostingRegressor(n_estimators=500, random_state=42)
model4.fit(X_train, y_train)
y_pred4 = model4.predict(X_val)
rmse4 = root_mean_squared_error(y_pred4, y_val)
rmse4
```
